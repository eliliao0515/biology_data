```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
from sklearn import preprocessing
```

```python
bmi = pd.read_csv('bmi.csv', index_col=False)
weight = pd.read_csv('weight.csv', index_col=False)
waistline = pd.read_csv('waistline.csv', index_col=False)
```

```python
#預留給資料前處理的地方
```

```python
bmilst = []
for j in bmi:
    bmilst.extend(list(bmi[j]))
```

```python
weightlst = []
for j in weight:
    weightlst.extend(list(weight[j]))
```

```python
waistlinelst = []
for j in waistline:
    waistlinelst.extend(list(waistline[j]))
```

```python
lst=[]
day = 1
for j in range(len(bmilst)):
    lst.append([day, bmilst[j], weightlst[j], waistlinelst[j]])
    day+=1
```

```python
df = pd.DataFrame(lst, columns=['day','bodyfat(%)','weight(kg)','waistline(cm)'])
df.plot(x='day', legend=True)
```
